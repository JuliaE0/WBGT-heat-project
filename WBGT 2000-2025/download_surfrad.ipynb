```python
import pandas as pd
import pvlib
from datetime import datetime, timedelta
from pathlib import Path

output_dir = Path("surfrad")
output_dir.mkdir(exist_ok=True)

BASE_URL = "https://gml.noaa.gov/aftp/data/radiation/surfrad"
STATION = "Desert_Rock_NV"

for year in range(2013, 2026):   # 2000 to 2026, but currently set to start at where I left off

    yearly_data = []

    start = datetime(year, 1, 1)
    end = datetime(year + 1, 1, 1)

    current = start

    while current < end:

        yy = str(year)[2:]
        doy = current.timetuple().tm_yday

        filename = f"dra{yy}{doy:03d}.dat"
        url = f"{BASE_URL}/{STATION}/{year}/{filename}"

        try:
            df, meta = pvlib.iotools.read_surfrad(
                url,
                map_variables=True
            )

            yearly_data.append(df)

            print("Downloaded", filename)

        except Exception as e:
            print("Missing:", filename)

        current += timedelta(days=1)

    # Save after the entire year has been downloaded
    if yearly_data:

        yearly_df = pd.concat(yearly_data)

        yearly_df.to_csv(
            output_dir / f"desert_rock_surfrad_{year}.csv"
        )

        print(f"Saved {year}")

    else:
        print(f"No data found for {year}")
```
